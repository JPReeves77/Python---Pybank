```python
"""
FINANCIAL ANALYSIS
Find total number of months in the .csv file
Find the net total of Profit/Losses over the entire period
Find the average of the changes in the Profit/Losses over the entire period
Find the greatest  increase in profits - date and amount - over the entire period
Find the greatest decrease in losses - date and amount - over the entire period
"""
import csv
total_months = 0
total_profit = 0
monthly_changes = []
max_profit = 0
max_loss = 0
month_list=[]

with open('budget_data.csv', 'r') as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    header = next(data)
    #print(header)
    first_row=next(data)
    previous_month_data = int(first_row[1])
    #print(previous_month_data)
    total_profit = previous_month_data
    total_months = total_months + 1
    #print(f"Total Months : {total_months}")
    #print(f"Total Profit : {total_profit}") 
    month_list.append(first_row[0])
    for row in data:
        total_months += 1
        #print(f"Total Months : {total_months}")
        total_profit += int(row[1])
        #print(f"Total Profit : {total_profit}")
        profit_change = int(row[1]) - int(previous_month_data)
        #print(f"monthly_change : {profit_change}")
        monthly_changes.append(profit_change)
        #print(monthly_changes[0:-1])
#         print(*average_change)
        previous_month_data = int(row[1])
        #print(f"New previous_month_data : {previous_month_data}")
        month_list.append(row[0])

import numpy as np
#print(np.sum(monthly_changes)/len(monthly_changes))

cum_change=0
change_count=0
max_profit_index=0
max_loss_index=0
max_profit=0
max_loss=0

# for each_change in monthly_changes: 
for i in range(len(monthly_changes)): 
    each_change=monthly_changes[i]
    cum_change=cum_change+each_change # cum_change+monthly_changes
    change_count=change_count+1
    if each_change>max_profit: 
        max_profit=each_change
        max_profit_index=i
    if each_change<max_loss: 
        max_loss=each_change
        max_loss_index=i

        # each_change=monthly_changes[0]
# cum_change=cum_change+each_change

# each_change=monthly_changes[1]
# cum_change=cum_change+each_change

# each_change=monthly_changes[2]
# cum_change=cum_change+each_change

average_change=cum_change/change_count
#print(average_change)

msg = "Financial Analysis\n------------------\n"
print(msg)
print(f"Total Months: {total_months}")
print(f"Total: ${total_profit}")
print(f"Average Change: ${average_change:0.2f}")
print(f"Greatest Increase in Profits: {month_list[max_profit_index+1]} (${max_profit})")
print(f"Greatest Decrease in Profits: {month_list[max_loss_index+1]} (${max_loss})")
```
